```python
import urllib.request
import requests
import urllib.parse
from bs4 import BeautifulSoup
import time
pageNum = 1
count = 1

lastPage = 1000

results = []
while pageNum < lastPage + 1:
    url = f'https://www.ekcm.co.kr/cust/initCustCenterReview.action?page_idx={pageNum}&'
    result =requests.get(url)
    bs_obj = BeautifulSoup(result.content, 'html.parser')
    tbody = bs_obj.find('tbody')
    number = tbody.findAll('span', {"class":"num"})
    title = tbody.findAll('td', {"class":"td_tit"})
    satisfy = tbody.findAll('span', {"class":"star_rate05"})
    date = tbody.findAll('span', {"class":"date"})
    content = tbody.findAll('span', {"class":"rcmt"})
    writer = tbody.findAll('td', {"class":"wrt"})

    for j in range(10):
        realnumber = number[j].text
        realtitle = title[j].text
        realsatisfy = satisfy[j].text
        realdate = date[j].text
        realcontent = content[j].text.replace("\t","").replace("\r","").replace("\n","")
        realwriter = writer[j].text
        data = [realnumber, realtitle, realsatisfy, realdate, realcontent, realwriter]
        results.append(data)
    time.sleep(0.5)
    pageNum += 1
    count += 1
```

```python
import pandas as pd
result_df = pd.DataFrame(results)
result_df.columns = ['번호', '상품', '만족도', '날짜', '내용', '작성자']
result_df.head()
```
