```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
driver = webdriver.Chrome(r'C:\Users\moonk\chromedriver.exe')
driver.get('https://cafe.naver.com/ArticleList.nhn?search.clubid=10789848&userDisplay=50&search.boardtype=L&search.specialmenutype=&search.totalCount=501&search.page=1')
```

```python
pageNum = 2
count = 1
i = input('몇페이지 크롤링 할까요? : ')
lastPage = int(i)
result = []
while pageNum < lastPage + 1:
    driver.get(f'https://cafe.naver.com/ArticleList.nhn?search.clubid=10789848&userDisplay=50&search.boardtype=L&search.specialmenutype=&search.totalCount=501&search.page={pageNum}')
    time.sleep(5)
    driver.switch_to.frame('cafe_main')
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    data = soup.findAll('div', {'class':'article-board m-tcol-c'})[1]

    category = data.findAll('div', {'class':'inner_name'})
    title = data.findAll('div', {'class':'inner_list'})
    name = data.findAll('div', {'class':'pers_nick_area'})
    date = data.findAll('td', {'class':'td_date'})
    view = data.findAll('td', {'class':'td_view'})
    
    for i in range(50):
        data = [category[i].text, title[i].text.replace("\n","").replace("   ","").replace("사진","").replace("new",""), name[i].text, date[i].text, view[i].text]
        result.append(data)
    pageNum += 1
    count += 1
```

```python
import pandas as pd
result_df = pd.DataFrame(result)
result_df.columns = ['category', 'title', 'name', 'date', 'view']
result_df
```

```python
result_df.to_excel('한울벗채식나라.xlsx')
```
